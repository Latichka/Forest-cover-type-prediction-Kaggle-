# Forest cover type prediction (Kaggle)

- **Цель**: предсказать тип лесного покрова на площади 30x30 кв.м по четырем районам дикой природы, расположенные в Национальном лесу Рузвельта на севере Колорадо. <br>
Всего 7 типов лесного покрова: 
1. Ель/Пихта
2. Сосна обыкновенная
3. Сосна Пондероза
4. Тополь/ива
5. Осина
6. Дугласовая пихта
7. Круммгольц
- **Данные**: данные предоставлены информационной системы ресурсов региона Лесной службы США (USFS)и Геологической службы США и USFS. Предсказание необходимо оуществить по картографичеким признакам: расстояние до подземных вод, близость воды, расстояние до проезжей части, индекс затемнения холмов, расстояние до точек пожара, тип почвы
- **Вывод** : таким образом, необходимо решить задачу многоклассовой классификации


- **План**:
1. Проанализировать полученные данные:
    - посмотреть распределение признаков, их описательные статистики (распределение, диперсия, средние значения и т.д.)
2. Подготовка данных:
    - разделить на категориальные и количественные значения;
    - удалить ненужные признаки;
    - обработать пропуски;
    - удалить дубликаты;
    - очистить данные: обработать выбросы, удалить явно ошибочные наблюдения;
    - проверить признаки на сильную корреляцию;
    - если необходимо, то создать новые фичи(признаки);
    - проверю дисбаланс классов;
3. Разделение датасета на тренировочную, валидационную и тестовую выборки;
4. Масштабирование признаков;
5. Обучение/выбор наилучшей модели.

Для оценки обученных моделей решено ипользовать метрику F1

```python
# Выделение признаков и таргета
x = data.drop('Cover_Type', axis= 1)
y = data['Cover_Type']
```

```python
# Раделение датасета на три части: тренировочная, валидационная, тестовая
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
x_train, x_cv, y_train, y_cv = train_test_split(x_train, y_train, test_size = 0.25)
```

Нейросеть
***

```python
#реализация простой нейресетевую архитектуру
class Mymodule (nn.Module):
    def __init__(self):
        super().__init__ ()
        self.fc1 = nn.Linear(50,128)
        self.fc2 = nn.Linear(128,64)
        self.fc3 = nn.Linear(64,7)
    
    def forward (self, x):
        x = self.fc1(x)
        x = nn.Sigmoid()(x)
        x= self.fc2(x)
        x= nn.Sigmoid()(x)
        x= self.fc3(x)
        return x
        
```

```python
model = Mymodule()
```

```python
#функция для цикл обучения нейросети
def train_model(x_tensor,y_tensor):
    for epoch in range(n_epochs):
        idx = torch.randperm(x_tensor.shape[0])
        x_t = x_tensor[idx].view(x_tensor.size())
        y_t = y_tensor[idx].view(y_tensor.size())
        l = 0
        avg_loss=0
        for i in range(int(x_tensor.shape[0]/batch_size)):
            x = x_t[l:l+batch_size]
            y = y_t[l:l+batch_size]
            l = l+batch_size
            optim.zero_grad()
            logits = model(x)
            loss = loss_fn(logits, y )
            loss.backward()
            optim.step()
            avg_loss+=loss.item()
        avg_loss = avg_loss/(len(y_train)/batch_size)
        print('loss: %f' %avg_loss)
    return model
        
```

```python
#обучение по немасштабированным данным:
start = time.time()
model_net = train_model(x_tensor,y_tensor)
stop = time.time()
```
